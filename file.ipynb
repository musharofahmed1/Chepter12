# File

```python
# write mode
file = open('f1.txt', mode='w')
file.write('Line4')
file.writelines(['\nLine5', '\nLine6'])
file.close()
```

```python
# append mode
file = open('f1.txt', mode='a')
file.write('\nLine7')
file.writelines(['\nLine8', '\nLine9'])
file.close()
```

```python
# xclusive mode
file = open('f1.txt', mode='x')
file.write('\nLine7')
file.writelines(['\nLine8', '\nLine9'])
file.close()
```

```python
# read mode
file = open('f1.txt', mode='r')
data = file.read(8)
print(data)
file.close()
```

```python
# read mode
file = open('f1.txt', mode='r')
data = file.readlines(6)
print(data)
file.close()
```

```python
# read bytes mode
file = open('f1.txt', mode='rb')
data = file.read(8)
print(data)
file.close()
```

```python
# read write mode
file = open('f1.txt', mode='r+')
data = file.read(8)
print(data)
file.write('\nLine11')
file.close()
```

```python
# read mode
file = open('f1.txt', mode='r')
file.write('hfdhss')
file.close()
```

```python
# read mode
file = open('f1.txt', mode='a')
file.read()
file.close()
```

```python
# Checkers
file = open('f1.txt', mode='a')
print('Readable:', file.readable())
print('Writable: ', file.writable())
print('Closed: ', file.closed)
file.close()
print('After Closed: ', file.closed)
```

# with

```python
with open('f1.txt', mode='a') as file:
    print('Readable:', file.readable())
    print('Writable: ', file.writable())
    file.write('Frim with')
    print('Closed: ', file.closed)
print('After Closed: ', file.closed)
```

```python
with open('f1.txt', mode='r+') as file:
    file.write('Write')
    d = file.read()
    print(d)
```
