## try and except

```python
res = 0.0
try:
    num = int(input("Dividend: "))
    div = int(input('Divisor: '))
    res = num/div
except Exception as e:
    print(e)

print(res)
```

## Specific Exception Handling

```python
res = 0.0
try:
    num = int(input("Dividend: "))
    div = int(input('Divisor: '))
    res = num/div
except ZeroDivisionError as z:
    res = num/1
    print(z)
except ValueError as v:
    res = num/2
    print(v)

print(res)
```

## Finally and else

```python
res = 0.0
try:
    num = int(input("Dividend: "))
    div = int(input('Divisor: '))
    res = num/div
except ZeroDivisionError as z:
    res = num/1
    print(z)
except ValueError as v:
    res = num/2
    print(v)
finally:
    print(res)
```
