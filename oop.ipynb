```python
# Creating Class
class Product:
    # Class or GLobal Attributes
    platform = 'AMAZON'

    # Constructor
    def __init__(self, pid: int, title: str, price: float) -> None:
        self.pid = pid
        self.title = title
        self.price = price
        self.code = 121

    # Methods
    def getInfo(self) -> str:
        return (self.pid,self.title, self.price, self.code)

    def getDiscount(self) -> float:
        return self.price - (self.price * (15/100))

    def __repr__(self) -> str:
        return f'Product(pid={self.pid}, title={self.title})'
```

```python
# Creating Object From The class
p1 = Product(243, 'IPHONE', 120000.00)
p2 = Product(243, 'IPHONE X', 150000.00)
print(p1)
```

# Getter, Setter, Deleter

```python
class Product:
    platform = 'AMAZON'

    def __init__(self, pid: int, title: str, price: float) -> None:
        self.pid = pid
        self.title = title
        self.price = price
        self.__code = 352323     # Private attribute

    @property           #getter
    def code(self) -> int:
        return self.__code

    @code.setter
    def code(self, new) -> None:
        self.__code = new

    @code.deleter
    def code(self) -> None:
        del self.__code

    def getDiscount(self) -> float:
        return self.price - (self.price * (15/100))

    def __repr__(self) -> str:
        return f'Product(pid={self.pid}, title={self.title})'
```

```python
p1 = Product(243, 'IPHONE', 120000.00)
p2 = Product(243, 'IPHONE X', 150000.00)
```

```python
p1.code             # calling getter
```

```python
p1.code = 904548     # calling setter
p1.code
```

```python
del p1.code
```

```python
del p1.price
```
